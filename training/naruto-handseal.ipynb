```python
import os
from roboflow import Roboflow

DATASET_VERSION = 2

api_key = os.getenv("ROBOFLOW_API_KEY", "")

#from kaggle_secrets import UserSecretsClient
#user_secrets = UserSecretsClient()
#secret_value_0 = user_secrets.get_secret("ROBOFLOW_API_KEY")

if not api_key:
    raise ValueError("Missing ROBOFLOW_API_KEY. Add it in Kaggle Secrets.")

rf = Roboflow(api_key=api_key)
project = rf.workspace("naruto-handseal").project("naruto-handsign")
version = project.version(DATASET_VERSION)
dataset = version.download("yolo26")

data_yaml_path = f"{dataset.location}/data.yaml"
print("Data downloaded to:", data_yaml_path)
```
